```python
class LogisticRegression:
  def __init__(self, x, y):
    self.x = x
    self.y = y
    self.n = len(x)
    self.a = 0
    self.b = 0
    self.alpha = 0.01
    self.epoch = 100
    self.cost_list = []

  def sigmoid(self, x):
    return 1/(1 + np.exp(-x))

  def train(self):
    for i in range(self.epoch):
      Z = self.a*self.x + self.b
      A = self.sigmoid(Z)

      # cost function
      cost = -(1/self.n)*np.sum( self.y*np.log(A) + (1-self.y)*np.log(1-A))

      # Gradient Descent

      da = (1/self.n)*np.dot(A-self.y, self.x)
      db = (1/self.n)*np.sum(A - self.y)

      self.a =self.a- self.alpha*da
      self.b = self.b-self.alpha*db

      # Keeping track of our cost function value
      self.cost_list.append(cost)

      print("cost after ", i, "iteration is : ", cost)
```
