```python

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
Normal=x['annualSalary'].values.reshape(-1,1)
x['annualSalary']=scaler.fit_transform(Normal)
Normal=x['grossPay'].values.reshape(-1,1)
x['grossPay']=scaler.fit_transform(Normal)
```

```python
salary_train, salary_test = train_test_split(x,train_size=0.7, random_state=42)
```

```python
salary_train
```

```python
salary_test
```

```python

y_train = salary_train.pop('annualSalary')

X_train = salary_train



y_test = salary_test.pop('annualSalary')

X_test = salary_test
```

```python
lr = LinearRegression()

salary_reg = lr.fit(X_train,y_train)
```

```python
salary_reg.score(X_train,y_train)
```

```python
salary_reg.score(X_test,y_test)
```

```python
result=lr.predict(X_test)
```

```python
mean_squared_error(y_test,lr.predict(X_test))
```

```python
mean_absolute_error(y_test,lr.predict(X_test))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 9))
plt.plot(result)

plt.show()
```

```python
plt.figure(figsize=(10,5))
y_train_pred = salary_reg.predict(X_train)
res = y_train - y_train_pred
sns.distplot(res)
plt.xlabel('Residuals')
plt.title("Residual Analysis",fontsize=20)
plt.show()
```

```python
sns.scatterplot(x=res,y=y_train_pred)
plt.xlabel('Residuals')
plt.title("Residual Analysis",fontsize=20)
plt.show()
```
